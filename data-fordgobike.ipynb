# Definição de função e aplicação de função em uma Série

```python
def retorna_estado(cidade):
    if cidade == 'Rio de Janeiro':
       return 'RJ'
    elif cidade == 'São Paulo':
         return 'SP'
    elif cidade == 'Fortaleza':
       return 'CE'
    elif cidade == 'Goiânia':
       return 'GO'
    else:
      return 'GO'
```

```python
cidades = pd.Series(['Rio de Janeiro','São Paulo','Fortaleza','Goiânia'])
uf = []
for cidade in cidades:
    uf.append(retorna_estado(cidade))
uf    
```

```python
cidades = pd.Series(['Rio de Janeiro','São Paulo','Fortaleza','Goiânia'])
cidades.apply(retorna_estado)
```
